```python
import pyvista as pv

file_path = "/PHShome/yl535/project/python/sam_3d/sam-3d-objects/results/FIVES/gaussians/1_A.ply"

# PyVista automatically detects if it's a mesh or point cloud
mesh = pv.read(file_path)

# Plotting
# 'color' sets the default color, 'point_size' is for point clouds
mesh.plot(color='cyan', point_size=5, show_edges=True)
```
